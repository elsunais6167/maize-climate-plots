```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('y30.csv')
data.head()
```

```python
year = data['Year']
temp = data['Annual-Mean']
a_temp = data['5-yr-Smooth']

plt.figure(figsize=(10,5))

plt.plot(year, temp, label='Annual Mean')
plt.plot(year, a_temp, label='5-yr-Smooth')

plt.title('30 Years Annual Mean Temperature of Nigeria - By El-Sunais (Data Source: CCKP, 2022)')
plt.ylabel('Temperature (deg C)')
plt.xlabel('Date')

plt.grid(linestyle=':')
plt.legend(loc='upper left')
plt.show()
```
